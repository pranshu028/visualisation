```python
# Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap.
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Generate a synthetic dataset with correlated features
np.random.seed(0)  # For reproducibility
num_samples = 100

# Creating correlated data
feature_1 = np.random.normal(0, 1, num_samples)
feature_2 = 2.5 * feature_1 + np.random.normal(0, 1, num_samples)
feature_3 = -1.5 * feature_1 + np.random.normal(0, 1, num_samples)
feature_4 = np.random.normal(0, 1, num_samples)

# Combine the features into a DataFrame
data = pd.DataFrame({
    'Feature 1': feature_1,
    'Feature 2': feature_2,
    'Feature 3': feature_3,
    'Feature 4': feature_4
})

# Calculate the correlation matrix
corr_matrix = data.corr()

# Visualize the correlation matrix using a heatmap
fig, ax = plt.subplots(figsize=(8, 6))
cax = ax.matshow(corr_matrix, cmap='coolwarm')
fig.colorbar(cax)

# Set ticks and labels
ax.set_xticks(range(len(corr_matrix.columns)))
ax.set_yticks(range(len(corr_matrix.columns)))
ax.set_xticklabels(corr_matrix.columns)
ax.set_yticklabels(corr_matrix.columns)

# Rotate the tick labels for better visibility
plt.xticks(rotation=45)

# Add titles
plt.title('Correlation Matrix Heatmap', pad=20)

# Display the plot
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Generate the dataset
np.random.seed(20)
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}
df = pd.DataFrame(data)

# Pivot the data to get the sales matrix
sales_matrix = df.pivot_table(values='Sales', index='Day', columns='Month', aggfunc='sum').fillna(0)

# Create a heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(sales_matrix, annot=True, fmt="d", cmap='YlGnBu', cbar=True)

# Add titles and labels
plt.title('Sales Variation Across Different Months and Days')
plt.xlabel('Month')
plt.ylabel('Day')

# Display the plot
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Generate the dataset for sales data
np.random.seed(20)
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}
df = pd.DataFrame(data)

# Pivot the data to get the sales matrix
sales_matrix = df.pivot_table(values='Sales', index='Day', columns='Month', aggfunc='sum').fillna(0)

# Create a heatmap using Matplotlib
fig, ax = plt.subplots(figsize=(12, 8))
cax = ax.matshow(sales_matrix, cmap='YlGnBu')

# Add color bar
fig.colorbar(cax)

# Set ticks and labels
ax.set_xticks(range(len(sales_matrix.columns)))
ax.set_yticks(range(len(sales_matrix.index)))
ax.set_xticklabels(sales_matrix.columns)
ax.set_yticklabels(sales_matrix.index)

# Rotate the tick labels for better visibility
plt.xticks(rotation=45)

# Add titles
plt.title('Sales Variation Across Different Months and Days', pad=20)
plt.xlabel('Month')
plt.ylabel('Day')

# Display the plot
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Generate the dataset
np.random.seed(25)
data = {
    'Country': ['USA', 'Canada', 'UK', 'Germany', 'France'],
    'Population': np.random.randint(100, 1000, 5),
    'GDP': np.random.randint(500, 2000, 5)
}
df = pd.DataFrame(data)

# Create a bubble chart
plt.figure(figsize=(10, 6))
scatter = plt.scatter(df['GDP'], df['Population'], s=df['Population']*10, alpha=0.5, c='blue')

# Add titles and labels
plt.title('Bubble Chart of Country Population and GDP')
plt.xlabel('GDP')
plt.ylabel('Population')

# Add text labels for each point
for i, row in df.iterrows():
    plt.text(row['GDP'], row['Population'], row['Country'], fontsize=12, ha='right')

# Display the plot
plt.show()
```

```python
Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Sine Wave', color='blue', linewidth=2)

plt.title('Sine Wave')
plt.xlabel('x')
plt.ylabel('y')

plt.legend()

plt.show()
```
